**4. Tuples (Creation, Modification and Access) :**

Create 1st tuple with values -> (10, 20, 30), 2nd tuple with values -> (40, 50, 60):

```python
first_tuple = (10,20,30)
second_tuple = (40,50,60)

#Concatenate the two tuples and store it in “t_combine”
t_combine = first_tuple + second_tuple

print("1st Tuple: ",first_tuple)
print("2nd Tuple: ",second_tuple)
print("Combined Tuple: ",t_combine)

# Repeat the elements of “t_combine” 3 times
t_repeat = t_combine * 3
print("3 Times of t_combine: ",t_repeat)

#Access the 3rd element from “t_combine”
third_element = t_combine[2]
print("Third element of t_combine is:",third_element)

#Access the first three elements from “t_combine”
first_three_element = t_combine[0:3]
print("First three elements: ",first_three_element)

# Access the last three elements from “t_combine”
last_three_element = t_combine[-3:]
print("Last three elements: ",last_three_element)
```
